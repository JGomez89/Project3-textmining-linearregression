```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("All_Reviews.csv")
df
```

```python
review_test = df.drop(columns = ['Unnamed: 0', 'ranking', 'category','review ID',
       'review name', "reviewer's name", "reviewer's profile", 'review rating',
       'review date', 'verified purchase',
       'vine customer review of free product', 'images', 'image 1', 'image 2',
       'image 3', 'image 4'], axis = 1 )
review_test['review text'] = review_test['review text'].apply(str)
test_df = review_test.groupby('filename')['review text'].apply(' '.join).reset_index()
test_df
```

```python
# Data preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
# Change to unicode
comments = test_df["review text"].values.astype("U")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(comments)
```

```python
k = 10
model = KMeans(n_clusters = k, init = 'k-means++',max_iter = 100, n_init = 1)
model.fit(features)
```

```python
test_df['category'] = model.labels_
```

```python
ct_clusters = pd.crosstab(test_df["filename"],
                 test_df["category"])
ct_clusters
```

```python
print("Words per catergory:")
cluster_rank = model.cluster_centers_.argsort()[:,::-1]
term = vectorizer.get_feature_names()

for i in range(k):
    print("Cluster %d:" % i)
    for j in cluster_rank[i, : 20]: #print out 10 feature terms of each cluster
        print (" %s'" % term[j])
    print("----------------")
```
